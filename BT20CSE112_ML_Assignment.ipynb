```python
import requests
from bs4 import BeautifulSoup
import re

# Function to fetch text from a URL
def fetch_text(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        return None

# URL of the page containing Shakespeare's works
url = 'https://www.gutenberg.org/files/100/100-0.txt'

# Fetching text from the URL
shakespeare_text = fetch_text(url)
sentence_string = ''
# Preprocessing the text
if shakespeare_text:
    # Remove metadata at the beginning of the text
    shakespeare_text = re.sub(r'.*START OF (THE|THIS) PROJECT GUTENBERG EBOOK.*', '', shakespeare_text, flags=re.IGNORECASE)
    # Remove metadata at the end of the text
    shakespeare_text = re.sub(r'.*END OF (THE|THIS) PROJECT GUTENBERG EBOOK.*', '', shakespeare_text, flags=re.IGNORECASE)
    # Remove punctuations
    shakespeare_text = re.sub(r'[^a-zA-Z\s]', '', shakespeare_text)
    # Remove extra spaces
    shakespeare_text = re.sub(r'\s+', ' ', shakespeare_text)
    # Convert to lowercase
    shakespeare_text = shakespeare_text.lower()

    # Print the preprocessed text
    # print(shakespeare_text[:50000])  # Print the first 1000 characters as an example
    sentence_string = shakespeare_text[:50000]  # Print the first 1000 characters as an example

else:
    print("Failed to fetch text from the URL.")
```
